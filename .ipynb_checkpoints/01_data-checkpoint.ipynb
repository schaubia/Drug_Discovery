# 01 — Data Loading & Exploration
**EGFR Kinase | ChEMBL Dataset**

Steps:
1. Load the EGFR data downloaded in 00_setup
2. Validate and clean SMILES with RDKit
3. Explore activity distribution and molecular properties
4. Create train / val / test splits
5. Save everything to ./data/ for the model notebooks

> Prerequisite: 00_setup.ipynb must have run — ./data/egfr_raw.csv must exist.

---
## STEP 1 — Load Config and Raw Data

```python
import json, os, warnings
import numpy as np
import pandas as pd
warnings.filterwarnings('ignore')

# Load config
with open('./config.json') as f:
    CONFIG = json.load(f)

SEED = CONFIG['random_seed']
np.random.seed(SEED)

print('Config loaded.')
print(f'  Target:      {CONFIG["target"]}')
print(f'  Random seed: {SEED}')
print()

# Load raw data
RAW_PATH = './data/egfr_raw.csv'
assert os.path.exists(RAW_PATH), 'ERROR: egfr_raw.csv not found — run 00_setup.ipynb first'

df_raw = pd.read_csv(RAW_PATH)
print(f'Raw data loaded: {len(df_raw):,} compounds')
print(f'Columns: {list(df_raw.columns)}')
print()
print(df_raw.head())
```

---
## STEP 2 — Clean and Validate SMILES with RDKit

```python
from rdkit import Chem
from rdkit.Chem import Descriptors, AllChem, Lipinski
from tqdm import tqdm

def get_mol_props(smiles):
    """Validate SMILES and return molecular properties. Returns None if invalid."""
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None
        return {
            'mol_weight':    round(Descriptors.MolWt(mol), 2),
            'logp':          round(Descriptors.MolLogP(mol), 2),
            'hbd':           Lipinski.NumHDonors(mol),
            'hba':           Lipinski.NumHAcceptors(mol),
            'tpsa':          round(Descriptors.TPSA(mol), 2),
            'rot_bonds':     Lipinski.NumRotatableBonds(mol),
            'heavy_atoms':   mol.GetNumHeavyAtoms(),
            'aromatic_rings': Descriptors.NumAromaticRings(mol),
        }
    except:
        return None

print('Validating SMILES and computing descriptors...')

props_list  = []
valid_flags = []

for smi in tqdm(df_raw['Drug'], desc='Processing'):
    props = get_mol_props(smi)
    props_list.append(props if props else {})
    valid_flags.append(props is not None)

df_raw['valid'] = valid_flags
n_invalid = sum(1 for v in valid_flags if not v)
print(f'\n  Invalid SMILES removed: {n_invalid}')

# Keep only valid rows
df_props = pd.DataFrame(props_list)
df_clean = pd.concat([df_raw.reset_index(drop=True), df_props], axis=1)
df_clean = df_clean[df_clean['valid']].drop(columns=['valid'])
df_clean = df_clean.dropna(subset=['Y']).reset_index(drop=True)

# Remove extreme outliers (IC50 > 100,000 nM — likely assay errors)
before = len(df_clean)
df_clean = df_clean[df_clean['Y'] < 100000].reset_index(drop=True)
print(f'  Outliers removed (IC50 > 100k nM): {before - len(df_clean)}')
print(f'\n  Clean dataset: {len(df_clean):,} compounds')
```

---
## STEP 3 — Exploratory Analysis

```python
import matplotlib.pyplot as plt

print('Dataset Overview')
print('='*45)
print(f'  Compounds:      {len(df_clean):,}')
print(f'  Unique SMILES:  {df_clean["Drug"].nunique():,}')
print()
print('IC50 (nM) statistics:')
print(df_clean['Y'].describe().round(2))
print()
print('Molecular weight statistics:')
print(df_clean['mol_weight'].describe().round(2))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 9))
fig.suptitle('EGFR Dataset — Exploratory Analysis', fontsize=13, fontweight='bold')

# IC50 raw distribution
ax = axes[0, 0]
ax.hist(df_clean['Y'], bins=60, color='steelblue', edgecolor='white', alpha=0.85)
ax.set_xlabel('IC50 (nM)')
ax.set_ylabel('Count')
ax.set_title('IC50 Distribution (raw)')

# Log IC50 distribution
log_y = np.log1p(df_clean['Y'])
ax = axes[0, 1]
ax.hist(log_y, bins=60, color='coral', edgecolor='white', alpha=0.85)
ax.set_xlabel('log(IC50 + 1)')
ax.set_title('Log IC50 Distribution (used for training)')

# Molecular weight
ax = axes[0, 2]
ax.hist(df_clean['mol_weight'], bins=50, color='mediumseagreen', edgecolor='white', alpha=0.85)
ax.axvline(500, color='red', linestyle='--', label='Lipinski MW=500')
ax.set_xlabel('Molecular Weight (Da)')
ax.set_title('Molecular Weight')
ax.legend(fontsize=8)

# LogP
ax = axes[1, 0]
ax.hist(df_clean['logp'], bins=50, color='mediumpurple', edgecolor='white', alpha=0.85)
ax.axvline(5, color='red', linestyle='--', label='Lipinski LogP=5')
ax.set_xlabel('LogP')
ax.set_title('LogP Distribution')
ax.legend(fontsize=8)

# Lipinski Rule of Five
ax = axes[1, 1]
ro5 = (
    (df_clean['mol_weight'] <= 500) &
    (df_clean['logp']       <= 5)   &
    (df_clean['hbd']        <= 5)   &
    (df_clean['hba']        <= 10)
)
pass_count = ro5.sum()
fail_count = (~ro5).sum()
ax.pie([pass_count, fail_count],
       labels=['Passes Ro5', 'Fails Ro5'],
       colors=['mediumseagreen', 'salmon'],
       autopct='%1.1f%%', startangle=90)
ax.set_title('Lipinski Rule of Five')

# LogP vs TPSA colored by activity
ax = axes[1, 2]
sc = ax.scatter(df_clean['logp'], df_clean['tpsa'],
                c=log_y, cmap='RdYlGn_r', alpha=0.4, s=6)
plt.colorbar(sc, ax=ax, label='log(IC50)')
ax.set_xlabel('LogP')
ax.set_ylabel('TPSA')
ax.set_title('LogP vs TPSA (colour = activity)')

plt.tight_layout()
os.makedirs('./results', exist_ok=True)
plt.savefig('./results/01_eda.png', dpi=150, bbox_inches='tight')
plt.show()
print('Plot saved to results/01_eda.png')
```

---
## STEP 4 — Known EGFR Drugs as Validation Anchors
These are drugs we know work — if our model scores them highly, it is working.

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

KNOWN_DRUGS = {
    'Gefitinib':   'COc1cc2ncnc(Nc3ccc(F)c(Cl)c3)c2cc1OCCCN1CCOCC1',
    'Erlotinib':   'C#Cc1cccc(Nc2ncnc3cc(OCCOC)c(OCCOC)cc23)c1',
    'Osimertinib': 'C=CC(=O)Nc1cc2c(Nc3ccc(F)c(Cl)c3)ncnc2cc1NC(=O)c1ccc(N(C)CCN(C)C)cc1',
    'Afatinib':    'CN(C)C/C=C/C(=O)Nc1cc2c(Nc3ccc(F)c(Cl)c3)ncnc2cc1OCC1=CN(C)C=C1',
}

print('Known EGFR Inhibitors — Validation Anchors')
print('='*55)
for name, smi in KNOWN_DRUGS.items():
    mol = Chem.MolFromSmiles(smi)
    if mol:
        mw  = Descriptors.MolWt(mol)
        lp  = Descriptors.MolLogP(mol)
        print(f'  {name:<15} MW={mw:.0f}  LogP={lp:.1f}')
    else:
        print(f'  {name:<15} invalid SMILES')

import json
with open('./data/known_drugs.json', 'w') as f:
    json.dump(KNOWN_DRUGS, f, indent=2)
print('\nSaved to data/known_drugs.json')
```

---
## STEP 5 — Log Transform Activity

```python
# IC50 spans many orders of magnitude — log transform stabilises regression
df_clean['Y_log'] = np.log1p(df_clean['Y'])

print('Activity transformation:')
print(f'  Original  mean={df_clean["Y"].mean():.1f}  std={df_clean["Y"].std():.1f}  skew={df_clean["Y"].skew():.2f}')
print(f'  Log+1     mean={df_clean["Y_log"].mean():.2f}  std={df_clean["Y_log"].std():.2f}  skew={df_clean["Y_log"].skew():.2f}')
print()
print('Skew should be much lower after log transform — better for regression.')
```

---
## STEP 6 — Train / Validation / Test Split

```python
from sklearn.model_selection import train_test_split

TEST_SIZE = CONFIG['test_size']   # 0.2
VAL_SIZE  = CONFIG['val_size']    # 0.1

# Split off test set first
df_trainval, df_test = train_test_split(
    df_clean, test_size=TEST_SIZE, random_state=SEED, shuffle=True
)

# Split remaining into train and val
val_fraction = VAL_SIZE / (1 - TEST_SIZE)
df_train, df_val = train_test_split(
    df_trainval, test_size=val_fraction, random_state=SEED, shuffle=True
)

total = len(df_clean)
print('Data splits:')
print(f'  Train:      {len(df_train):>6,}  ({len(df_train)/total*100:.0f}%)')
print(f'  Validation: {len(df_val):>6,}  ({len(df_val)/total*100:.0f}%)')
print(f'  Test:       {len(df_test):>6,}  ({len(df_test)/total*100:.0f}%)')
print(f'  Total:      {total:>6,}')
print()
print('Activity mean per split (should be similar):')
print(f'  Train:      {df_train["Y_log"].mean():.3f}')
print(f'  Validation: {df_val["Y_log"].mean():.3f}')
print(f'  Test:       {df_test["Y_log"].mean():.3f}')
```

---
## STEP 7 — Save Processed Data

```python
os.makedirs('./data', exist_ok=True)

df_train.to_csv('./data/train.csv',      index=False)
df_val.to_csv('./data/val.csv',          index=False)
df_test.to_csv('./data/test.csv',        index=False)
df_clean.to_csv('./data/full_clean.csv', index=False)

print('Saved:')
print(f'  data/train.csv       {len(df_train):,} rows')
print(f'  data/val.csv         {len(df_val):,} rows')
print(f'  data/test.csv        {len(df_test):,} rows')
print(f'  data/full_clean.csv  {len(df_clean):,} rows')
print(f'  data/known_drugs.json')
print()
print('✓ Data pipeline complete — ready for 02_baselines.ipynb')
```
